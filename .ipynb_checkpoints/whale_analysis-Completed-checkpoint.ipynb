```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
import seaborn as sns
%matplotlib inline
```

# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:
1. whale_returns.csv
2. algo_returns.csv
3. sp500_history.csv

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")
whale_returns = pd.read_csv(whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale_returns.head()
```

```python
# Count nulls
whale_returns.isnull().sum()
```

```python
# Drop nulls
whale_returns = whale_returns.dropna().copy()
whale_returns
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
algo_returns = pd.read_csv(algo_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_returns
```

```python
# Count nulls
algo_returns.isnull().sum()
```

```python
# Drop nulls
algo_returns = algo_returns.dropna().copy()
algo_returns
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
sp500_returns = pd.read_csv(sp500_history_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp500_returns.sort_index(inplace=True)
sp500_returns
```

```python
# Check Data Types
whale_returns.dtypes
```

```python
algo_returns.dtypes
```

```python
sp500_returns.dtypes
```

```python
sp500_returns["Close"] = sp500_returns["Close"].str.replace("$", "")
sp500_returns["Close"]
```

```python
# Fix Data Types
sp500_returns["Close"] = sp500_returns["Close"].astype("float")
sp500_returns.dtypes
```

```python
# Calculate Daily Returns
# Calculate the daily return using the 'shift()' function

sp500daily_returns = sp500_returns.pct_change()
sp500daily_returns.head()
```

```python
# Drop nulls
sp500daily_returns.dropna().copy()
```

```python
# Rename Column
sp500daily_returns.rename(columns={'Close':'S&P 500'}, inplace=True)
sp500daily_returns.tail()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
combined_returns = pd.concat([whale_returns, algo_returns, sp500daily_returns], axis="columns", join="inner")
combined_returns
```

# Portfolio Analysis

In this section, you will calculate and visualize performance and risk metrics for the portfolios.

## Performance

Calculate and Plot the daily returns and cumulative returns. Does any portfolio outperform the S&P 500? 

```python
# Plot daily returns
combined_returns.plot(figsize=(20,10), title='Daily Returns')
```

```python
# Calculate cumulative returns
cumulative_returns = (1 + combined_returns).cumprod()
cumulative_returns.head()
```

```python
# Plot cumulative returns
cumulative_returns.plot(figsize=(20,10), title='Cumulative Returns')
```

#### ANSWER: As of end of the data timeframe, only Berkshire Hathaway Inc and Algo 1 portfolios have outperformed the S&P 500

## Risk

Determine the _risk_ of each portfolio:

1. Create a box plot for each portfolio. 
2. Calculate the standard deviation for all portfolios
4. Determine which portfolios are riskier than the S&P 500
5. Calculate the Annualized Standard Deviation

```python
# Box plot to visually show risk
combined_returns.plot.box(figsize=(20,10), title='Portfolio Risk')
```

#### ANSWER: Berkshire Hathaway has the largest spread, and Algo 1 has the smallest spread.

```python
# Daily Standard Deviations
# Calculate the standard deviation for each portfolio. 
# Which portfolios are riskier than the S&P 500?
daily_std = combined_returns.std()
daily_std
```

```python
# Determine which portfolios are riskier than the S&P 500
daily_std = daily_std.sort_values(ascending=False)
daily_std
```

#### ANSWER: Berkshire Hathaway and Tiger Global Management have higher risk than the S&P 500

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = daily_std * np.sqrt(252)
annualized_std
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
combined_returns.rolling(window=21).std().plot(figsize=(20,10), title='21-Day Rolling Standard Deviation')
```

#### ANSWER: Portfolio risk generally follows the S&P 500, with the exception of Berkshire Hathaway, which periodically showed more pronounced risk increases than the market, and Tiger Global Management, which sowed a few peaks in risk that were not consistent with fluctuations in the market or the rest of the portfolios. On the other hand, Paulson & Co Inc., was consistently below the risk level of the S&P 500 and often had the lowest risk of all portfolios.

```python
# Construct a correlation table
correlation = combined_returns.corr()
correlation
```

```python
# Use the `heatmap` function to visualize correlations
sns.heatmap(correlation, vmin=-1, vmax=1)
```

#### ANSWER: Algo 1 showed the highest correlation with the S&P 500

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)

# Calculate 30-day rolling covariance of Algo 1 vs. S&P 500
rolling_covariance = combined_returns['Algo 1'].rolling(window=21).cov(combined_returns['S&P 500'])

# Calculate 30-day rolling variance of S&P 500
rolling_variance = combined_returns['S&P 500'].rolling(window=21).var()

# Calculate 30-day rolling beta of Algo 2 and plot the data
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Algo 1 Beta')
```

#### ANSWER: Algo 1 seems highly sensitive to movements in the S&P 500

```python
# Calculate a rolling window using the exponentially weighted moving average. 
combined_returns.ewm(span=21).mean(window=21).plot(figsize=(20,10), title = 'Exponentially Weighted Average')
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
sharpe_ratios = (combined_returns.mean() * 252) / (combined_returns.std() * np.sqrt(252))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios.plot(kind="bar", title="Sharpe Ratios", figsize=(20,10), fontsize=(13))
```

On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales? 
Type your answer here:   

##### ANSWER: Yes, Algo 1 outperforms both, 'the market' and the whales. And Algo 2 outperforms most whales, except Berkshire Hathaway, and does not outperform the S&P 500. As a combined portfolio, the Algos outperform 'the market' and the whales.

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Read the first stock
# Add 'Symbol' column
# Rename Index and 'Close' column
# Reset order of columns
goog_stock_csv = Path("Resources/GOOG_Stock_Prices.csv")
goog_stock = pd.read_csv(goog_stock_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
goog_stock['Symbol'] = 'GOOG'
goog_stock = goog_stock.rename(columns={'Close':'NOCP'}, index={'Date':'Trade DATE'})
goog_stock = goog_stock[['Symbol','NOCP']]
goog_stock.head()
```

```python
# Read the second stock
# Add 'Symbol' column
# Rename Index and 'Close' column
# Reset order of columns
amzn_stock_csv = Path("Resources/AMZN_Stock_Prices.csv")
amzn_stock = pd.read_csv(amzn_stock_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
amzn_stock['Symbol'] = 'AMZN'
amzn_stock = amzn_stock.rename(columns={'Close':'NOCP'}, index={'Date':'Trade DATE'})
amzn_stock = amzn_stock[['Symbol','NOCP']]
amzn_stock.head()
```

```python
# Read the third stock
# Add 'Symbol' column
# Rename Index and 'Close' column
# Reset order of columns
jpm_stock_csv = Path("Resources/JPM_Stock_Prices.csv")
jpm_stock = pd.read_csv(jpm_stock_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
jpm_stock['Symbol'] = 'JPM'
jpm_stock = jpm_stock.rename(columns={'Close':'NOCP'}, index={'Date':'Trade DATE'})
jpm_stock = jpm_stock[['Symbol','NOCP']]
jpm_stock.head()
```

```python
# Concatenate all stocks into a single DataFrame
my_portfolio = pd.concat([goog_stock, amzn_stock, jpm_stock], axis="rows", join="inner")
my_portfolio.head()
```

```python
# Reset the index
my_portfolio = my_portfolio.reset_index()
my_portfolio.head()
```

```python
# Pivot so that each column of prices represents a unique symbol
my_portfolio = my_portfolio.pivot_table(values="NOCP", index="Date", columns="Symbol")
my_portfolio.head()
```

```python
# Calculate returns for my portfolio
my_returns = my_portfolio.pct_change()
my_returns.head()
```

```python
# Drop Nulls
my_returns.dropna(inplace=True)
my_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
weighted_returns = my_returns.dot(weights)
weighted_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
all_returns1 = pd.concat([combined_returns, weighted_returns], axis="columns", join="inner")
all_returns1
```

```python
# Rename index
all_returns1.rename_axis("Trade DATE", axis='index', inplace=True)
all_returns1.head()
```

```python
# Rename columns
all_returns1.columns = ['SOROS FUND MANAGEMENT LLC', 'PAULSON & CO INC', 'TIGER GLOBAL MANAGEMENT LLC', 'BERKSHIRE HATHAWAY INC', 'ALGO 1', 'ALGO 2', 'S&P 500', 'MY PORTFOLIO']
all_returns1.head()
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
all_returns1.dropna(inplace=True)
all_returns1.head()
```

## Re-run the performance and risk analysis with your portfolio to see how it compares to the others

```python
# Calculate risk
all_daily_std = all_returns1.std()
all_daily_std
```

```python
# Calculate volatility
volatility = all_returns1.std() * np.sqrt(252)
volatility
```

```python
# Plot rolling Standard Deviation
all_returns1.rolling(window=21).std().plot(figsize=(20,10), title='21-Day Rolling Standard Deviation')
```

```python
# Annualized Sharpe Ratios
all_sharpe_ratios = (all_returns1.mean() * 252) / (all_returns1.std() * np.sqrt(252))
all_sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
all_sharpe_ratios.plot(kind="bar", figsize=(20,10))
```

```python
# Create a correlation analysis
all_correlation = all_returns1.corr()
all_correlation
```

```python
# Use the `heatmap` function to visualize correlation of all returns
sns.heatmap(all_correlation, vmin=-1, vmax=1)
```

```python
# Calculate rolling covariance
rolling_covariance = all_returns1['MY PORTFOLIO'].rolling(window=30).cov(all_returns1['S&P 500'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 30-Day Covariance of Custom Portfolio Returns vs. S&P 500 Returns')
```

```python
# Calculate 30-day rolling variance of S&P 500
rolling_variance = all_returns1['S&P 500'].rolling(window=30).var()
rolling_variance.plot(figsize=(20, 10), title='Rolling 30-Day Variance of S&P 500 Returns')
```

```python
# Calculate 30-day rolling beta of MY PORTFOLIO and plot the data
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 30-Day Beta of MY PORTFOLIO')
```

#### ANSWER: While My Portfolio was overall riskier and more volatile than the market, the total Sharpe Ratio value and the down-trending value of the rolling 30-Day Beta showed that this is a portfolio moving in the right direction.
